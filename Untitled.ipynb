```python
Vaccine_data = pd.read_csv("country_vaccinations_by_manufacturer.csv")
data=pd.DataFrame(columns=['Country', 'Vaccine', 'Total_vaccine'])
for country in Vaccine_data["location"].unique():
    for vaccine in Vaccine_data["vaccine"].unique():
        filtered_data = Vaccine_data[(Vaccine_data['location'] == country) & (Vaccine_data['vaccine'] == vaccine)]
        total_count = filtered_data['total_vaccinations'].max()
        data = pd.concat([data, pd.DataFrame({'Country': [country], 'Vaccine': [vaccine], 'Total_vaccine': [total_count]})], ignore_index=True)
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# searching for all punctuation
data_2.loc[:, 'punctuation'] = data_2['text'].apply(lambda x: ''.join(set(re.findall(r'[^\w\s]+', str(x)))))

# replace punctuation & with and
data_2.loc[:, 'reviewText']= data_2['text'].apply( lambda x: str(x).replace('&', 'and')) 

# Preprocessed Data
Pre_data_2 = data_2[['text']]

wordnet_lemmatizer = WordNetLemmatizer()  # Initialize WordNet Lemmatizer

def cleaningText(text):
    text = re.sub("[^a-zA-Z0-9]", " ", text)  # Remove Punctuation
    text = re.sub("[0-9]+", "", text)  # Remove Numbers
    text = [wordnet_lemmatizer.lemmatize(word.lower()) for word in word_tokenize(text) if word.lower() not in stopwords.words('english')]
    return " ".join(text)

Pre_data_2.loc[:, 'clean_review'] = Pre_data_2['text'].apply(lambda x: cleaningText(str(x)))

# Tokenize clean_review column
#Pre_data.loc[:,'tokenized_review'] = Pre_data['clean_review'].apply(word_tokenize)
```

```python
import panel as pn
import hvplot.pandas
import plotly.express as px

# Load Plotly extension
pn.extension('plotly')

# Define Panel widget for selecting a country
unique_countries = ['All'] + review_data['Country'].unique().tolist()
country_select = pn.widgets.Select(name='Country', options=unique_countries)

# Function to aggregate data across all countries or for a specific country
def aggregate_data(country='All'):
    if country == 'All':
        total_data = review_data.sum(numeric_only=True).to_frame().T
    else:
        total_data = review_data[review_data['Country'] == country]
    return total_data

# Function to plot aggregated data
def plot_aggregated_data(country):
    aggregated_data = aggregate_data(country)
    return aggregated_data.hvplot.bar(
        title=f'Total COVID-19 Cases for {country}',
        xlabel='Measure',
        ylabel='Count',
        rot=45
    )

# Define Panel layout for first main section
country_dashboard = pn.Column(
    pn.bind(plot_aggregated_data, country=country_select)
)

# Function to create Tabulator widget
def create_tabulator_widget(country='All'):
    co19_pipeline = aggregate_data(country)
    co19_table = co19_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size=10, sizing_mode='stretch_width')
    return co19_table

# Define Panel layout for second main section
tabulator_dashboard = pn.Column(
    pn.bind(create_tabulator_widget, country=country_select)
)

# Create Plotly Express plots
# Assuming 'Vaccination' is your DataFrame containing vaccination data
vaccine_map = px.choropleth(Vaccination, 
                            locations='iso_code', 
                            color='vaccines',
                            scope='world')  # Specifying the scope as 'world'
vaccine_map.update_layout(height=300, margin={"r":0,"t":0,"l":0,"b":0}, showlegend=False)

# Assuming 'total_vaccinations_by_country_vaccines' is your DataFrame containing the total vaccinations by country and vaccines
fig = px.bar(data, 
             x='Vaccine', 
             y='Total_vaccine', 
             title='Total Vaccinations by Vaccine',
             labels={'vaccines': 'Vaccines', 'total_vaccinations': 'Total Vaccinations'})


import matplotlib.pyplot as plt

def plot_sentiment_counts(sentiment_counts):

    plt.figure(figsize=(4, 3))
    sentiment_count_plot = sentiment_counts.plot(kind='bar', color=['red', 'blue', 'yellow', 'green'])
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.title('Count of Each Sentiment Category')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent clipping labels
    return plt.gcf()

# Example usage:
# Assuming sentiment_counts is already defined
# plot = plot_sentiment_counts(sentiment_counts)
# plt.show()  # Call this later to display the plot


# Convert the matplotlib plot to a Panel object
#sentiment_count_pane = pn.pane.Matplotlib(plt.gcf(), tight=True)


# Generate word cloud
def generate_word_cloud(all_text):
    # Specifying stop words
    stop_words = set(STOPWORDS)

    # Generating the word cloud
    word_cloud = WordCloud(width=600, height=300, stopwords=stop_words).generate(all_text)

    # Plotting the word cloud
    plt.figure(figsize=(6, 3), facecolor='k')
    plt.title('Tweets for Covid_19')
    plt.imshow(word_cloud)
    plt.axis('off')  # Turn off axis
    plt.tight_layout(pad=0)
    # Don't close the figure here
    return plt.gcf()  # Return the figure object

# Get all text data
all_text = ' '.join(Pre_data_2.clean_review.tolist())

# Generate word cloud plot
word_cloud_plot = generate_word_cloud(all_text)
plot_sentiment_counts_plot =plot_sentiment_counts(sentiment_counts)

# Create the main layout
main_layout = pn.Column(
    pn.Row(
        pn.Row(country_dashboard, tabulator_dashboard, sizing_mode='stretch_width', margin=(0, 25)),
        sizing_mode='stretch_width'
    ),
    pn.Row(
        pn.pane.Plotly(vaccine_map), 
        pn.pane.Plotly(fig),
        sizing_mode='stretch_width', margin=(0, 25)
    ),
    pn.Row(
    word_cloud_plot,plot_sentiment_counts_plot,sizing_mode='stretch_width', margin=(0, 25),
    # Add your components or plots here

)
# Define FastListTemplate
    

)

# Define the bottom layout
#bottom_layout = 
template = pn.template.FastListTemplate(
    title='Covid 19 dashboard', 
    sidebar=[
        pn.pane.Markdown("# Covid 19"), 
        pn.pane.Markdown("#### Today I'm excited to share with you a project that focuses on understanding COVID-19 cases and vaccination efforts around the world.As we all know, the COVID-19 pandemic has had a profound impact on our lives, with millions of cases reported worldwide. Through careful analysis of data, we've been able to gather insights into the spread of the virus, including confirmed, death, and recovered cases."), 
        pn.pane.JPG('23312_lores.jpg', sizing_mode='scale_both'),
        pn.pane.Markdown("## Settings"),   
        country_select,
    ],
    main=main_layout,
    #bottom=bottom_layout,
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)

# Display the dashboard
template.servable()
```
